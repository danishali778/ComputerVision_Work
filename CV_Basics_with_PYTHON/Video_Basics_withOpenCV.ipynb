```python
import cv2
import time
```

```python
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Could not open video capture.")

else:

  width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
  height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
  ## FOR WINDOWS *'DIVX'
  writer = cv2.VideoWriter('myvideo.mp4',cv2.VideoWriter_fourcc(*'DIVX'),20,(width,height))

  while True:
    ret,frame = cap.read()
    # OPERATION (DRAWING)
    writer.write(frame)
    gray_scale = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    cv2.imshow('frame',gray_scale)

    if(cv2.waitkey(0) & 0xFF == ord('q')):
      break

  cap.release()
  writer.release()
  cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture("gdrive/My Drive/Datasets/DATA/")

if(cap.isOpened() ==False):
  print("Failed to open File ")

while cap.isOpened():
  ret,frame = cap.read()  # It is originally reads video frames  very fast,we can delay it by using time module

  if(ret == True):
    time.sleep(1/20) # It means 20 frames per second
    cv2.imshow('frame',frame)
    if(cv2.waitKey(0) & 0xFF == ord('q')):
      break
    else:
      break

cap.release()
cv2.destroyAllWindows()
```

## Drawing on Video


```python
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Could not open video capture.")

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# TOP LEFT CORNER
x = width  // 2
y = height // 2

w = width//4
h = height //4

# BOTTOM RIGHT CORNER


while True:
  ret,frame = cap.read()
  cv2.rectangle(frame,(x,y),(x+w,y+h),color=(0,0,255),thickness=4)
  cv2.imshow('frame',frame)

  if (cv2.waitKey(0) & 0xFF = ord('q')):
    break

cap.release()
cv2.destroyAllWindows()
```

```python


## CALLBACK FUNCTION RECTANGLE
def draw_rectangle(event,x,y,flags,params):
  global pt1,pt2,topLeft_clicked,bottomRight_clicked

  if (event == cv2.EVENT_LBUTTONDDOWN):
    # RESET THE RECTANGLE

    if topLeft_clicked and bottomRight_clicked:
      pt1 = (0,0)
      pt = (0,0)
      topLeft_clicked = False
      bottomRight_clicked = False

    if topLeft_clicked == False:
      pt1 = (x,y)
      topLeft_clicked = True
    elif bottomRight_clicked == False:
      pt2 = (x,y)
      bottomRight_clicked = True



## GLOBAL VARIABLES
pt1 = (0,0)
pt2 = (0,0)

topLeft_clicked = False
bottomRight_clicked = False

## CONNECT TO THE CALLBACK
cap = cv2.VideoCapture(0)
cv2.namedWindow('Test')
cv2.setMouseCallback('Test',draw_rectangle)



while True:
  ret,frame = cap.read()
  ## DRAWING ON THE FRAME BASED OFF THE GLOBAL VARIABLES

  if topLeft_clicked:
    cv2.circle(frame,center=pt1,radius = 5,color=(0,0,255),thickness = -1)

  if topLeft_clicked and bottomRight_clicked:
    cv2.rectangle(frame,pt1,pt2,color=(0,0,255),thickness = -1)
  cv2.imshow('Test',frame)

  if (cv2.waitKey(0) & 0xFF = ord('q')):
    break

cap.release()
cv2.destroyAllWindows()


```
