```python
## Direct Drawning with mouse
# we use callbacks to connect images with mouse
# --> when we connect with specific call backs they take some extra arguments which i have provided in draw_circle function.

img = np.zeros(shape=(512,512,3),dtype = np.int8)

#################
## FUNCTION #####
#################

# x and y is mouse's current position.
def draw_circle(event,x,y,flags,param):
  if(event == cv2.EVENT_LBUTTONDOWN):
    cv2.circle(img,(x,y), 100,(0,255,0),-1)

  elif event == cv2.cv2.EVENT_RBUTTONDOWN:
    cv2.circle(img,(x,y), 100,(255,0,0),-1)


# These lines below connecting the above function to image below in the while loop.
cv2.namedWindow(winname = 'image')
cv2.setMouseCallback('image',draw_circle) # here parameters  should match the image to be shown on call back.

#####################################
## SHOWNING IMAGE WITH OPENCV #######
#####################################

while True:
  cv2.imshow("image",img)
  if(cv2.waitKey(20) & 0xFF == 27):
    break
cv2.destroyAllWindows()
```

```python

# VARIABLES
# False while mouse button is up.
drawing = False  # make it true if mouse is pressed down( meaning we are drawing something)
ix = -1
iy = -1

# FUNCTION

def draw_rectangle(event,x,y,flags,params):
  global  ix,iy,drawing

  if event == cv2.EVENT_LBUTTONDOWN:
    drawing = True
    ix,iy = x,y # get position of mouse
  elif event == cv2.EVENT_MOUSEMOVE:
    if(drawing == True):
      # in rectangle we need two points where we started to draw(ix,iy) and where we are going which is x and y
      cv2.rectangle(img,(ix,iy),(x,y),(0,255,0),-1)

  elif event == cv2.EVENT_LBUTTONUP:
    drawig = False
    cv2.rectangle(img,(ix,iy),(x,y),(0,255,0),-1)



# SHOWING THE IMAGE

img = np.zeros((512,512,3))

cv2.namedWindow(winname='image')
cv2.setMouseCallback('image',draw_rectangle)

while True:
  cv2.imshow('image',img)

  if cv2.waitKey(1) & 0xFF = 27:
    break

cv2.destroyAllWindows()
```
